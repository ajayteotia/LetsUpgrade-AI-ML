## Parametric and Non-Parametric Test

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

dataset = pd.read_excel("dataset_yes.xlsx")
dataset
```

## Wilcoxon Test

### H0 = There is no significant difference between Attrition rate and Employee's Distance from home ###
### H1 = There is a significant difference between Attrition rate and Employee's Distance from home  ###

```python
from scipy.stats import wilcoxon
stats,p = wilcoxon(dataset.Attrition,dataset.DistanceFromHome)
print(stats,p)
```

### Value of p is less than 0.05. Hence, we will reject the null hypothesis ###

## Friedman Test

### H0 = There is no significant difference between Attrition rate and Employee's Education and Age ###
### H1 = There is a significant difference between Attrition rate and Employee's Education and Age  ###

```python
from scipy.stats import friedmanchisquare
stats,p = friedmanchisquare(dataset.Attrition,dataset.Age,dataset.Education)
print(stats,p)
```

### Value of p is less than 0.05. Hence, we will reject the null hypothesis ###

## Chi Square Test

### H0 = There is no dependency between Attrition rate and Employee's Department ###
### H1 = There is dependency between Attrition rate and Employee's Department  ###

```python
from scipy.stats import chi2_contingency
chitable = pd.crosstab(dataset2.Attrition,dataset2.Department)
chitable
```

```python
stats,p,dof,expected = chi2_contingency(chitable)
print(stats,p)
```

### Value of p is less than 0.05. Hence, we will reject the null hypothesis ###

## One Sample T Test

### H0 = There is no significant difference between the mean of Employee's Distance from home against the population mean ###
### H1 = There is a significant difference between the mean of Employee's Distance from home against the population mean  ###

## Two Sample Paired Test

### H0 = There is no significant difference between the mean of Employee's Education and Education Field ###
### H1 = There is a significant difference between the mean of Employee's Education and Education Field  ###

```python
from scipy.stats import ttest_rel
stats,p=ttest_rel(dataset.Education,dataset.EducationField)
print(stats,p)
```

## Two Sample T Test

### H0 = There is no significant difference between the mean of Distance from home between Male employee and Female employee ###
### H1 = There is a significant difference between the mean of Distance from home between Male employee and Female employee  ###

```python
from scipy.stats import ttest_ind
stats,p=ttest_ind(dataset.Gender,dataset.DistanceFromHome)
print(stats,p)
```

### Value of p is less than 0.05. Hence, we will reject the null hypothesis ###

## One Way ANOVA Test 

### H0 = There is no significant impact on Attrition rate because of Employee's Education ###
### H1 = There is a significant impact on Attrition rate because of Employee's Education  ###

```python
import statsmodels.api as sm
from statsmodels.formula.api import ols
model = ols('Attrition~C(Education)',dataset1).fit()
oneway = sm.stats.anova_lm(model,typ=2)
print(oneway)
```

### Value of p is more than 0.05. Hence, we will accept the null hypothesis ###

## Two Way ANOVA Test

### H0 = There is no significant impact on Attrition rate because of Last year training time and Years at company ###
### H1 = There is a significant impact on Attrition rate because of Last year training time and Years at company ###

```python
model = ols('Attrition~C(TrainingTimesLastYear)+C(YearsAtCompany)',dataset1).fit()
twoway = sm.stats.anova_lm(model,typ=2)
print(twoway)
```

### Value of p is less than 0.05. Hence, we will reject the null hypothesis ###
